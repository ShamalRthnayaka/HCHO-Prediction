```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Reading the dataset
col_mat_nuw_data = pd.read_csv("Datasets/col_mat_nuw_output.csv", header=None)
col_mat_nuw_data
```

```python

# Define the new column names
column_names = ['HCHO Reading', 'Location', 'Current Date', 'Next Date']

# Rename the columns
col_mat_nuw_data.columns = column_names
```

```python
# Basic overview of the dataset
print(col_mat_nuw_data.head())
```

```python
# Data Schema
print(col_mat_nuw_data.info())
```

```python
# Checking for null values
print(col_mat_nuw_data.isnull().sum())
```

```python
# Plotting histogram for 'HCHO Reading' column
plt.hist(col_mat_nuw_data['HCHO Reading'], bins=20, color='blue', alpha=0.7)
plt.xlabel('HCHO Reading')
plt.ylabel('Frequency')
plt.title('Histogram of HCHO Reading')
plt.grid(True)
plt.show()
```

```python
# Replace null values in 'HCHO Reading' column with the mean
hcho_mean = col_mat_nuw_data['HCHO Reading'].mean()
col_mat_nuw_data['HCHO Reading'].fillna(hcho_mean, inplace=True)
```

```python
# Check if there are any null values left
print(col_mat_nuw_data.isnull().sum())
```

```python
# Plotting histogram after imputation
plt.hist(col_mat_nuw_data['HCHO Reading'], bins=20, color='blue', alpha=0.7)
plt.xlabel('HCHO Reading')
plt.ylabel('Frequency')
plt.title('Histogram of HCHO Reading (After Imputation)')
plt.grid(True)
plt.show()
```

```python
# Remove duplicates
col_mat_nuw_data_no_duplicates = col_mat_nuw_data.drop_duplicates()
```

```python
# Loop through each column and print unique values
for col_name in col_mat_nuw_data_no_duplicates.columns:
    unique_values = col_mat_nuw_data_no_duplicates[col_name].unique()
    print(f"Unique values in column '{col_name}':")
    for value in unique_values:
        print(value)
    print()
```

```python
# Create box plot
plt.figure(figsize=(8, 6))
sns.boxplot(x=col_mat_nuw_data_no_duplicates['HCHO Reading'], color='skyblue')
plt.title('Box Plot of HCHO Reading')
plt.xlabel('HCHO Reading')
plt.show()
```

```python
# Filter out rows with 'HCHO Reading' above 0.0006
col_mat_nuw_data_filtered = col_mat_nuw_data_no_duplicates[col_mat_nuw_data_no_duplicates['HCHO Reading'] <= 0.0006]

# Show the DataFrame after removing outliers
print(col_mat_nuw_data_filtered)

# Create box plot after filtering
plt.figure(figsize=(8, 6))
sns.boxplot(x=col_mat_nuw_data_filtered['HCHO Reading'], color='skyblue')
plt.title('Box Plot of HCHO Reading (Outliers Removed)')
plt.xlabel('HCHO Reading')
plt.show()
```

```python
# Save the DataFrame to a new CSV file
col_mat_nuw_data_filtered.to_csv("Filtered_col_mat_nuw_output.csv", index=False)
```
